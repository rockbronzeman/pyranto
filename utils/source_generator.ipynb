```python
working_file = "../keywords_tr_formatted.txt"
q = "\""
translator_dict = "trans_map"
functions_set = "funcs_set"
```

```python
## left to right
funcs_set = set()
outkeys = ""
outfunc = ""
exceptions_header = -1
print(f"{translator_dict} = " + "{")

with open(working_file) as f:    
    header = True ## header at top
    for line in f:
        exceptions_header += 1
        
        if header: ## skip
            if line.strip() == "exceptions": ## exception handled elsewhere
                break
            header = False
            continue

        line = line.split()
        if not line: ## empty line is preceeding header indicator
            header = True
            continue

        ## line is bijection representation
        if line[1] == "SKIP": continue
            
        if line[1][-1] == "*": ## remove asterix
            line[1] = line[1][:-1]
            
        if line[0][-1] == ")": ## remove call parentheses
            line[0], line[1] = line[0][:-2], line[1][:-2]
            outfunc += q + line[1] + q + ",\n"
            
        outkeys += q + line[0] + q + ":" + q + line[1] + q + ",\n"

print(outkeys[:-2]) ## get rid of comma       
print("}\n") ## extra newline

print(f"{functions_set} = " + "{")
print(outfunc[:-2]) ## get rid of comma
print("}")

print("\n" + str(exceptions_header))
```
