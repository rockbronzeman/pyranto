```python
infile  = "../keywords_tr.txt"
outfile = "../keywords_tr_formatted.txt"
left_width = 30
safety = False
```

```python
if not safety:
    with open(infile) as inf:
        with open(outfile, "w") as outf:
            
            header = True ## header at top
            for line in inf:
                if header: ## write header
                    header = False
                    outf.write(line)
                    continue
                
                line = line.split()
                if not line: ## empty line is preceeding header indicator
                    header = True
                    outf.write("\n")
                    continue

                available = left_width - len(line[0]) ## to format RHS nicely
    
                outf.write(line[0] + " "*available + " ".join(line[1:]) + "\n")
else: print("Safety on")
```
